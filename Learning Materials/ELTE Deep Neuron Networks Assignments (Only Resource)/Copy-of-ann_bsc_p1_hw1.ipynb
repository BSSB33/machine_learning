**Katalógus 2.**

A katalógus rendszer nagy sikerén felbuzdulva az egyetem a szolgáltatás továbbfejlesztésén gondolkodik. Sajos eddig a szolgáltatás népszerűsége miatt nem volt lehetőség leállítani azt, viszont a COVID-19 járvány miatt eljött a várva várt idő. 

Kezdetben egy külsős fejlesztői csapatot terveztek felkérni, de sajnos uniós támogatás hiánya miatt rád hárult a feladat. A kar vezetése azt reméli, hogy az új eszköz segítségével tovább növelhető a hallgatók tanulmányi átlaga.

Mivel a régi kódbázis refaktorálásra szorul, inkább kezdd nulláról a fejlesztést. Az új szoftver bejelentkezések formájában kapja meg az órán jelen levő hallgatók neptun kódját (6 karakter hosszú sztring). A neptun kódok mellé el kell, hogy tárolja a bejelentkezésre használt eszköz MAC-címét is (sztring).

Készíts egy **Catalog2** osztályt, amelynek az alábbi publikus metódusai vannak:

* **is_valid_neptun(neptun)**: Igazat ad vissza, ha pontosan 6 karakter hosszú a kapott neptun kód.
* **login(neptun, mac_address)**: Ha helyes a megadott neptun kód (6 karakter hosszú), akkor bejelentkezteti a hallgatót, eltárolva annak neptun kódját, mac címét.
* **get_students()**: Visszaadja a lelkes hallgatók neptunkódjainak a rendezett listáját, akik legalább háromszor jelentkeztek be a félév során (minden neptun kód egyszer szerepelhet a listában).
* **get_mac_addresses(neptun)**: Visszaadja az adott neptun kóddal rendelkező hallgató általt használt mac címek rendezett listáját (minden cím egyszer szerepelhet a listában).
* **get_number_of_logins(neptun)**: Visszaadja, hogy összesen hányszor jelentkezett be az adott neptun kóddal rendelkező hallgató.
* **get_loners()**: Azon lelkes (legalább háromszor bejelentkezett) hallgatók neptunkódjainak rendezett listájával tér vissza, akik kizárólag olyan eszközökről jelentkeztek be, melyeket mások nem használtak (minden mac cím egyszer szerepelhet a listában).

```

# Implement your solution below

class Catalog2():
    lista = []
    def is_valid_neptun(self, neptun):
        return len(neptun) == 6
    def login(self, neptun, mac_address):
        if self.is_valid_neptun(neptun):
            self.lista.append(neptun)
            self.lista.append(mac_address)
    def get_students(self):
        student_list = list(set([x for x in self.lista if self.lista.count(x) >= 3 and self.is_valid_neptun(x)]))
        return student_list
    def get_mac_addresses(self, neptun):
        mac_list = []
        for i in range(0, len(self.lista) - 1):
            if self.lista[i] == neptun:
                mac_list.append(self.lista[i + 1])
        solution = list(set(mac_list))
        return solution
        return mac_list
    def get_number_of_logins(self, neptun):
        return self.lista.count(neptun)
    def get_loners(self):
        loners = []
        students = self.get_students()
        for i in range(0, len(self.lista) - 1):
            if self.lista[i] in students and self.lista.count(self.lista[i + 1]) == 1:
                loners.append(self.lista[i])
        return loners


# Implement your solution above

# Tests
c = Catalog2()
func_list = ['is_valid_neptun', 'login', 'get_students', 'get_mac_addresses', 'get_number_of_logins', 'get_loners']
if any([getattr(c, fun, None) for fun in func_list]):
  assert c.get_students() == []
  assert c.get_loners() == []

  assert not c.is_valid_neptun('alma')
  c.login('alma', 42)
  assert c.get_number_of_logins('alma') == 0
  assert c.get_mac_addresses('alma') == []
  assert c.get_loners() == []

  assert c.is_valid_neptun('qweqwe')
  c.login('qweqwe', 42)
  assert c.get_students() == []
  assert c.get_number_of_logins('qweqwe') == 1
  assert c.get_mac_addresses('qweqwe') == [42]
  assert c.get_loners() == []

  c.login('qweqwe', 42)
  c.login('qweqwe', 24)
  assert c.get_students() == ['qweqwe']
  assert c.get_number_of_logins('qweqwe') == 3
  assert c.get_mac_addresses('qweqwe') == [24, 42]
  assert c.get_loners() == ['qweqwe']

  c.login('catalo', 24)
  assert c.get_students() == ['qweqwe']
  assert c.get_number_of_logins('catalo') == 1
  assert c.get_mac_addresses('catalo') == [24]
  assert c.get_number_of_logins('qweqwe') == 3
  assert c.get_mac_addresses('qweqwe') == [24, 42]
  assert c.get_loners() == []
  print('OK')
```

### B

**Accuracy/binary cross-entropy**

Ebben a feladatban két, a mély hálók tanítása és kiértékelése során gyakran használt függvényt kell implementálnod.

* **categorical_accuracy(y_pred, y_true)**: 
Multiclass klasszifikáció esetén a mintaelemeink címkéjét k kategória egyikébe becsüljük (például kutya, macska, papagáj, stb.). Ha a becsléshez neuronhálót használunk, tipikusan mintaelemenként egy-egy k elemű, valószínűségeket tartalmazó vektort becslünk, melyek a mintaelem egyes kategóriákba tartozásának valószínűségét reprezentálják. Ha szeretnénk ezekből a vektorokból megmondani, hogy egy-egy mintaelem melyik kategóriába tartozik a legnagyobb valószínűség szerint, akkor elég megmondani minden egyes vektorban a maximális elem indexét. A becslés pontosságának (accuracy) kiértékeléséhez kiszámolandó, hogy a tesztadatbázis elemeire mekkora arányban egyezik a becsült kategóriacímke és az igazi kategóriacímke (`y_true`).
Implementáld ezt a függvényt úgy, hogy az, az `y_pred` paraméterként kapja meg a becsült valószínűségeket (egy (m, k) méretű tömböt, ahol $m$ a mintaelemek és $k$ a kategóriák száma), valamint az `y_true` paraméterként kapja meg az igazi kategóriacímkéket (egy (m,) méretű tömböt) és visszadja, annak az arányát, hogy a becsült kategória címkéje egyezik az igazi kategóriacímkével.

* **binary_cross_entropy(y_pred, y_true)**: A függvény kiszámolja a  logisztikus költséget (logistic loss), melyet bináris klasszifikáció megoldásához használunk költségként. Azért, hogy elkerüljük a $\log(0)$ értékeket, az implementáció során adj hozzá egy $10^{-9}$-es (epszilon) értéket a logaritmus függvény belsejében megadott kifejezéselhez. A függvény az `y_pred` paraméterként várja a becsült valószínűségeket minden mintaelemre, egy (m,) méretű tömbben, ahol $m$ a mintaelemek száma. Az `y_true` paraméterben az igazi valószínűségeket kapja meg a függvény (0, vagy 1, attól függően melyik kategóriába tartozik az adott címke) egy (m,) méretű tömbben. A függvény egy számmal tér vissza, a mintaelemek felett átlagolt logisztikus költséggel, melynek képlete alább látható:

$$ -\frac{1}{m}\sum_{j=1}^m y \cdot \log(\hat{y}) + (1 - y) \cdot \log(1 - \hat{y})$$

ahol `y_pred` tárolja az $\hat{y}$ és `y_true` tárolja az $y$ értékeket az $m$ mintaelemre.

**Kikötés:**  Az implementációt vektoros módon, NumPy-ban, ciklusok és egyéb, annak megfelelő Python konstrukciók használata nélkül kell elkészíteni. További részletek a notebook végén.

```
import numpy as np

# Implement your solution below

def categorical_accuracy(y_pred, y_true):
    maxes = np.argmax(y_pred, axis=1)
    bools = np.equal(maxes, y_true)
    np.count_nonzero(bools)
    solution = np.count_nonzero(bools) / y_true.size
    return solution

def binary_cross_entropy(y_pred, y_true):
    constant = -1 / y_pred.size
    solution = np.sum(y_true * np.log(y_pred + 10**-9) + (1 - y_true) * np.log(1 - y_pred + 10**-9))
    solution *= constant
    return solution

# Implement your solution above

# Tests
y_pred = np.array([[.1, .5, .4], [0., .1, .9], [.2, .6, .2]])
y_true = np.array([0, 2, 1])
assert abs(categorical_accuracy(y_pred, y_true) - 0.666) < 0.01

y_pred = np.array([0.])
y_true = np.array([0.])
assert abs(binary_cross_entropy(y_pred, y_true) + 1e-9) < 1e-10

y_pred = np.array([1.])
y_true = np.array([1.])
assert abs(binary_cross_entropy(y_pred, y_true) + 1e-9) < 1e-10

y_pred = np.array([0., .1, .9])
y_true = np.array([0., 0., 1.])
assert abs(binary_cross_entropy(y_pred, y_true) -0.0702403) < 1e-4

y_pred = np.array([.1, .5, .4, 0., .1, .9, .2, .6, .2])
y_true = np.array([1., 0., 0., 0., 0., 1., 0., 1., 0.])
assert abs(binary_cross_entropy(y_pred, y_true) - 0.519376) < 1e-5
print('OK')
```

### C/3.

Írj egy **ChessboardKnights** nevű osztályt, ami egy sakktáblát reprezentál. A sakktábla üres kezdetben. A táblára huszárokat (ló figura) helyezünk fel, de olyan helyre nem rakhatunk fel huszárt, amit már üt legalább egy fenn levő huszár.

A következő tagfüggvényekkel kell, hogy rendelkezzen:
* `place(pos)`: ahol `pos` egy kételemű tuple, 0 és 7 közti (inkluzív) egész számokat kap és egy logikai értéket ad vissza, amivel megadja, hogy adott pozícióra (x_idx, y_idx) helyezhető-e újabb huszár és ha igen, fel is helyezi azt.
* `is_placable(pos)`: hasonló a `place(pos)` függvényhez, de nem módosítja a tábla állapotát.
* `is_any_place_left()` függvény megadja, hogy létezik-e még mező, amire felhelyezhető huszár.
* `get_n_free_places_in_col(x_idx)` függvény megadja, hogy az `x_idx` indexű oszlopban (0..7) hány olyan mező van, ahol még nincs huszár és amit még nem üt huszár.
* `get_n_free_places_in_row(y_idx)` függvény megadja, hogy az `y_idx` indexű sorban (0..7) hány olyan mező van, ahol még nincs huszár és amit még nem üt huszár.

**Kikötés:**  Az implementációt vektoros módon, NumPy-ban, ciklusok és egyéb, annak megfelelő Python konstrukciók használata nélkül kell elkészíteni. További részletek a notebook végén.

```
import numpy as np

# Implement your solution below
class ChessboardKnightsNpy:
  def __init__(self):
    self.board = np.zeros((8,8))
  def is_placable(self, pos):
    return self.board[pos] == 0
  def place(self, pos):
    if self.is_placable(pos):
      self.board[pos] = 1
      ones = np.array([-1, 1])
      twos = np.array([-2, 2])
      targets = np.transpose([np.tile(ones, len(twos)), np.repeat(twos, len(ones))])
      targets = np.concatenate((np.flip(targets, 1), targets))
      targets += pos
      bools1 = np.array(np.logical_and(targets[:,0] >= 0, targets[:,1] >= 0))
      bools2 = np.array(np.logical_and(targets[:,0] < 8, targets[:,1] < 8))
      bools = np.logical_and(bools1, bools2)
      targets = targets[bools]
      targets = np.concatenate(([pos], targets))
      self.board[targets[:,0], targets[:,1]] = 1
      return True
    return False

  def is_any_place_left(self):
    return np.sum(self.board) < 64

  def get_n_free_places_in_col(self, x_idx):
    return 8 - np.sum(self.board[x_idx,:])
  def get_n_free_places_in_row(self, y_idx):
    return 8 - np.sum(self.board[:,y_idx])

# Implement your solution above
# olyan cellák, az x és y távolságok összege 3 és egyik sem 0
# np.where [x-ek], [y-ok]
# Tests
cb = ChessboardKnightsNpy()
func_list = ['place', 'is_placable', 'is_any_place_left', 'get_n_free_places_in_col', 'get_n_free_places_in_row']
if any([getattr(cb, fun, None) for fun in func_list]):
  assert cb.get_n_free_places_in_col(5) == 8
  assert cb.get_n_free_places_in_col(0) == 8
  assert cb.is_any_place_left()
  assert cb.place((3,7))
  assert not cb.is_placable((2,5))
  assert not cb.place((2,5))
  assert cb.place((3,6))
  assert cb.place((3,4))
  assert cb.place((0,0))
  assert not cb.place((2,1))
  assert not cb.is_placable((1,2))
  assert cb.is_any_place_left()
  assert cb.place((6,5))
  assert cb.get_n_free_places_in_row(2) == 5
  assert cb.place((6,3))
  assert cb.place((6,1))
  assert cb.place((4,3))
  assert cb.place((2,3))
  assert cb.is_placable((1,0))
  assert cb.place((1,0))
  assert cb.get_n_free_places_in_row(7) == 4
  assert cb.place((3,3))
  assert cb.place((2,0))
  assert cb.place((2,7))
  assert cb.place((4,7))
  assert cb.place((6,7))
  assert cb.get_n_free_places_in_col(3) == 1
  assert cb.place((0,3))
  assert cb.place((0,5))
  assert cb.place((0,7))
  assert cb.place((3,0))
  assert cb.place((5,0))
  assert cb.place((6,0))
  assert cb.place((7,0))
  assert cb.place((7,4))
  assert cb.is_any_place_left()
  assert cb.place((7,6))
  assert cb.get_n_free_places_in_col(7) == 0
  assert not cb.is_any_place_left()
  print("OK")
```
