```python
def findDog(sentence):
    return 'dog' in sentence.lower().split(" ")

print(findDog('Is there a dog here?'))
print(findDog('Is there a DOG here?'))
print(findDog('Is there a cat here?'))
```

```python
def countDog(string):
    return sum(map(lambda x : (x == "dog") == 1, string.split())) 

print(countDog("Is there dog dog a dog here?"))
```

## Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2

plt.plot(x, y, 'r')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
```

```python
#fig = plt.figure(figsize=(8, 4), dpi = 100)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
fig.savefig("filename.png", dpi= 200)
```

## Exercises

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y, 'b')
ax.set_title("title")
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x, y, 'r')
ax2.plot(x, y, 'r')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])

ax1.plot(x, z, 'b')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')


ax2.plot(x, y, 'b')
ax2.set_ylim(30,55)
ax2.set_xlim(20,22)
ax2.set_title('zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'b--', linewidth=3.00)
axes[1].plot(x, z, 'r', linewidth=3.00)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,2))
axes[0].plot(x, y, 'b--', linewidth=3.00)
axes[1].plot(x, z, 'r', linewidth=3.00)
```
